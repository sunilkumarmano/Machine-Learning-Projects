#### 3.7 Dropping the column "Unnamed: 32"

```python
df.drop(columns = "Unnamed: 32", axis = 1, inplace = True)
df.head()
```

#### 3.8 Dropping the column "id"

The column "id" is considered to be the identity column of the patient. So, it can be removed as it will not be helpful for analysis in anyway.

```python
df.drop(columns = "id", axis = 1, inplace = True)
df.head()
```

#### 3.9 Converting the target variable from String to Numerical value

As the target variable column "diagnosis" is in String value, we can convert it into a numerical value.

As per the research,

- Malignant is a dangerous Tumor and leads to cancer.
- Benign Tumor cells are not cancerous.

So, lets mark

- Malignant as 1
- Benign as 0

```python
df["diagnosis"] = df["diagnosis"].map({'M' : 1, 'B' : 0})
df.head()
```

#### 3.10 Seperating the predictor variables as X

```python
X = df.loc[:, df.columns != 'diagnosis']
X.head()
```

#### 3.11 Seperating the target variable as Y

```python
Y = df.loc[:, df.columns == "diagnosis"]
Y.head()
```

### 4. Model Building

#### 4.1 Seperating the Train and Test split values for prediction

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 1)
```

#### 4.2 Checking the shape of X_train and Y_train values

```python
print(X_train.shape)
print(Y_train.shape)
```

#### 4.3 Checking the shape of X_test and Y_test values

```python
print(X_test.shape)
print(Y_test.shape)
```

#### 4.4 Building the Random Forest Classifier (Round 1)

```python
#Below two lines are used to ignore warnings produced by the IDE.
import warnings
warnings.filterwarnings("ignore")

from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state = 0)
rfc.fit(X_train, Y_train)
Y_pred = rfc.predict(X_test)
```

### 5. Model Evaluation

Let's evaluate the accuracy of the model by finding the accuracy score and confusion matrix.

#### 5.1 Checking the accuracy of the Model (Round 1)

```python
from sklearn.metrics import accuracy_score, confusion_matrix

print("Accuracy Score : ", accuracy_score(Y_pred, Y_test))
print("Confusion Matrix: \n", confusion_matrix(Y_pred, Y_test))
```

On seeing the above accuracy score we can conclude that our model can the Breast cancer with __94.5%__ accuracy.

Confusion matrix shows that only 8 out 143 predictions are wrong.

Let's try, if we can improve the accuracy of our model even more.

### 6. Model Improvement

As a first step in model improvement process we are going to tune the parameters given to the Random Forest Classifier as see the results.

Note that, in the Round 1 of Model development, though we have not given any paremeters explicitly, default parameters would taken by the compiler.

Now we are going to given the parameters explicitly and try if we can improve the model accuracy.

#### 6.1 Using RandomizedSearchCV for hyper parameter tuning (Round 2)

RandomizedSearchCV is used for hyperparameter tuning in model building process on various iterations by providing different input paremeters on each iterations.

Let's limit the number of iterations as 50.

For parameter values we can use the below ranges.

- "max_depth": range(2,5), 
- "min_samples_split": sp_randint(2, 11),
- "min_samples_leaf": sp_randint(1, 11),
- "bootstrap": [True, False],
- "n_estimators": [100, 400, 700, 1000, 1500],
- "criterion" : ["gini", "entropy"],
- "max_features": ['sqrt', 'log2', None]

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint as sp_randint

parameters = {
    "max_depth" : range(2,5),
    "min_samples_split" : sp_randint(2,11),
    "min_samples_leaf" : sp_randint(1,11),
    "bootstrap" : [True, False],
    "n_estimators" : [100, 400, 700, 1000, 1500],
    "criterion" : ["gini", "entropy"],
    "max_features" : ["sqrt", "log2", None]
}

randomizedSearchCV = RandomizedSearchCV(RandomForestClassifier(random_state = 0), 
                                        param_distributions = parameters, 
                                        n_iter = 50,
                                        n_jobs = -1)

randomizedSearchCV.fit(X_train, Y_train)
Y_pred1 = randomizedSearchCV.predict(X_test)
```

#### 6.2 Model evaluation after parameter tuning (Round 2)

```python
from sklearn.metrics import accuracy_score, confusion_matrix

print("Accuracy Score : ", accuracy_score(Y_pred1, Y_test))
print("Confusion Matrix : \n", confusion_matrix(Y_pred1, Y_test))
```

On seeing the above accuracy score we can conclude that our model accuracy score is improved from the Round 1 accuracy and it can predict the Breast cancer with __95.1%__ accuracy.

Confusion matrix shows that only __7 out 143__ predictions are wrong, which is a slight improvement.

Now, lets to find the principal components, i.e., the features that are critical for our prediction using Principal Component Analysis(PCA).

### 7. Applying Principal Component Analysis(PCA)

#### 7.1 Applying Standard Scaler for Predictor variables

Before applying the principal component analysis technique, we need to make the predictor or independent variables X to same scale.

So, we are applying Standard Scaler technique to bring the feature variables to same scale.

```python
from sklearn.preprocessing import StandardScaler

standardScaler = StandardScaler()
standardScaler.fit(X)
scaled_X = standardScaler.transform(X)
scaled_X
```

#### 7.2 Applying PCA and finding out the Principal Components

Now, lets find the number of principal components which gives maximum variance.

Based on that we can decide, how many principal components we need to have our prediction.

```python
from sklearn.decomposition import PCA

pca = PCA(random_state = 0)
pca.fit(scaled_X)
```

#### 7.3 Visualizing the variance of principal components

```python
plt.plot(pca.explained_variance_ratio_)
plt.xlabel("No. of features")
plt.ylabel("Variance of the features ")
plt.title("PCA Analysis")
plt.show()
```

From the visualization we can consider that variance is explained well by 11-12 features. 

After that the lines becomes flat, which means the other features are not much helpful in our prediction.

So, lets consider the 12 features for building our model and check the accuracy.

#### 7.4 Reducing the feature variables by applying PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 12, random_state = 0)
pca.fit(scaled_X)
pca_X = pca.transform(scaled_X)
pca_X.shape
```

#### 7.5 Splitting the data into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(pca_X, Y, test_size = 0.25, random_state = 0)
```

#### 7.6 Checking the size of X_train and Y_train values

```python
print(X_train.shape)
print(Y_train.shape)
```

#### 7.7 Checking the size of X_test and Y_test values

```python
print(X_test.shape)
print(Y_test.shape)
```

#### 7.8 Model Development with the principal components (Round 3)

Now, lets give the principal components as input to the randomizedSearchCV iterative model that we have developed before (Round 2) and check the accuracy score and confusion matrix.

```python
randomizedSearchCV.fit(X_train, Y_train)
Y_pred2 = randomizedSearchCV.predict(X_test)
```

#### 7.9 Model evaluation after providing principal componets to the model (Round 3)

```python
print("Accuracy score : ", accuracy_score(Y_pred2, Y_test))
print("Confustion Matrix : \n", confusion_matrix(Y_pred2, Y_test))
```

From the above evaluation, we can conclude that our 12 principal components are predicting the future values with only __90.9%__ accuracy and our accuracy score is reduced when compared to Round 2 evaluation and wrong predictions also increased from 7 to 13. So, let's leave out the Principal component analysis section for our predictions and end our predictions upto Round 2.

So, we have developed a model which can predict the Breast cancer with __95%__ accuracy for the future values.
